```python
df.loc[df['neighbourhood group'] == 'brookln', 'neighbourhood group'] = 'Brooklyn'
df.loc[df['neighbourhood group'] == 'manhatan', 'neighbourhood group'] = 'Manhattan'
```

```python
plt.figure(figsize=(20,10))
sns.catplot(data=df,x='neighbourhood group',kind='count')
plt.xticks(rotation=45)
```

```python
count_by_neighborhood = df.groupby('neighbourhood').count().sort_values('id',ascending=False)['id']
count_by_neighborhood
```

```python
df.loc[df['price'].isna(), 'price'] = '$0'
```

```python
df[['to_delete','actual_price']] = df['price'].str.split('$', expand=True)
```

```python
df['actual_price'].replace(',','', regex=True, inplace=True)
```

```python
df['actual_price'] = pd.to_numeric(df['actual_price'])
```

```python
df['actual_price'].dtype
```

```python
df.drop(['price','to_delete'],axis=1,inplace=True)
```

```python
df.head(5)
```

```python
filtered_price_data = df[df['actual_price'] != 0]
filtered_price_data.head(5)
```

```python
non_na_neighborhoods_groups = filtered_price_data[(filtered_price_data['neighbourhood group'] == 'Brooklyn') | (filtered_price_data['neighbourhood group'] == 'Manhattan') | (filtered_price_data['neighbourhood group'] == 'Queens') | (filtered_price_data['neighbourhood group'] == 'Bronx') | (filtered_price_data['neighbourhood group'] == 'Staten Island')]
```

```python
plt.figure(figsize=(20,3))
sns.boxplot(x=filtered_price_data['actual_price'])
```

```python
for i in non_na_neighborhoods_groups['neighbourhood group'].unique():
    temp = non_na_neighborhoods_groups[non_na_neighborhoods_groups['neighbourhood group'] == i]
    print(i,temp['actual_price'].mean())
```

```python
for i in non_na_neighborhoods_groups['neighbourhood group'].unique():
    temp = non_na_neighborhoods_groups[non_na_neighborhoods_groups['neighbourhood group'] == i]
    print(i,temp.mode()['actual_price'][0])
```

```python
for i in non_na_neighborhoods_groups['neighbourhood group'].unique():
    t = non_na_neighborhoods_groups[non_na_neighborhoods_groups['neighbourhood group'] == i]
    plt.figure(figsize=(10,5))
    sns.displot(t['actual_price'], kind='kde', height=5, aspect=3)
    plt.title(i)
```
